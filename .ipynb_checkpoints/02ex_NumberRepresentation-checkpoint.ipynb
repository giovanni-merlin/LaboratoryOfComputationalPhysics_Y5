1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
"""# an integer in decimal representation
a = 23

# its binary representation
a_bin = bin(a)
print('Binary representation of', a, ':', a_bin)

# its hexadecimal representation: convenient for data storage cause 16=2^4
a_hex = hex(a)
print('Hexadecimal representation of', a, ':', a_hex)

# converting back to integer
print('Decimal representation of', a_bin, ':', int(a_bin, 2))
print('Decimal representation of', a_hex, ':', int(a_hex, 16))

#int is in dec representation by default
#from bin to dec to hex
b_bin = 100001
b_list = list(map(int,str(b_bin))) #convert to list
"""

#x = input("choose a number: ")
x = 'a8'
# % is mod: returns the reminder of division
# // is the floor division

def conversion(x):
    #i = input("Choose your representation (bin, dec, hex): ")
    i = "hex"
    if (i == "bin"):
        
        x_bin_list = list(map(int,str(x))) #convert x to list
        #che differenza con fare solo list?
        x_bin_list.reverse()
        
        #qui converto in decimale
        x_dec = 0
        for i in range(0,len(x_bin_list),1): #range up to length-1
            x_dec = x_dec + x_bin_list[i]*2**i
        print("its decimal representation is:", x_dec)
        
        #qui converto in esa
        x_hex_list = []
        while (x_dec // 16 != 0):
            x_hex_list.append(x % 16)
            x_dec = x_dec // 16
        x_hex_list.append(x_dec % 16)   #per aggiungere il resto alla fine del while
        
        for i in range(0,len(x_hex_list)):
            if (x_hex_list[i] == 10):
                x_hex_list[i] = 'a'
            elif (x_hex_list[i] == 11):
                x_hex_list[i] = 'b'
            elif (x_hex_list[i] == 12):
                x_hex_list[i] = 'c'
            elif (x_hex_list[i] == 13):
                x_hex_list[i] = 'd'
            elif (x_hex_list[i] == 14):
                x_hex_list[i] = 'e'
            elif (x_hex_list[i] == 15):
                x_hex_list[i] = 'f'
            else:
                pass
          
        x_hex_list.reverse()
        print("its hexadecimal representation is:", x_hex_list)
        
        
    elif (i == "dec"):     

        x2 = x
        #qui devo convertire in binario
        
        x_bin_list2 = []
        while (x // 2 != 0):
            x_bin_list2.append(x % 2)
            x = x // 2
        x_bin_list2.append(1)   #per aggiungere 1 alla fine del while (è il resto)    
        x_bin_list2.reverse()
        x_bin_str = str(x_bin_list2)
        print("its binary representation is: ", x_bin_str)           
              
        #qui devo convertire in esa
        
        x_hex_list2 = []
        while (x2 // 16 != 0):
            x_hex_list2.append(x2 % 16)
            x2 = x2 // 16
        x_hex_list2.append(x2 % 16)   #per aggiungere il resto alla fine del while
        
        for i in range(0,len(x_hex_list2)):
            if (x_hex_list2[i] == 10):
                x_hex_list2[i] = 'a'
            elif (x_hex_list2[i] == 11):
                x_hex_list2[i] = 'b'
            elif (x_hex_list2[i] == 12):
                x_hex_list2[i] = 'c'
            elif (x_hex_list2[i] == 13):
                x_hex_list2[i] = 'd'
            elif (x_hex_list2[i] == 14):
                x_hex_list2[i] = 'e'
            elif (x_hex_list2[i] == 15):
                x_hex_list2[i] = 'f'
            else:
                pass
          
        x_hex_list2.reverse()
        print(x_hex_list2)
        
    elif (i == "hex"):
        #qui converto in binario: nibble by nibble
        
        x_bin_list = list(x) #convert x to list
        print(x_bin_list)
        for i in range(0,len(x_bin_list)):
            if (x_bin_list[i] == 'a'):
                x_bin_list[i] = '1010'
            elif (x_bin_list[i] == 'b'):
                x_bin_list[i] = '1011'
            elif (x_bin_list[i] == 'c'):
                x_bin_list[i] = '1100'
            elif (x_bin_list[i] == 'd'):
                x_bin_list[i] = '1101'
            elif (x_bin_list[i] == 'e'):
                x_bin_list[i] = '1110'
            elif (x_bin_list[i] == 'f'):
                x_bin_list[i] = '1111'
            else:
                x_bin_list[i] = int(x_bin_list[i])
                aux_list = [] #auxiliary list to append for every nibble
                while (x_bin_list[i] // 2 != 0):
                    aux_list.append(x_bin_list[i] % 2)
                    x_bin_list[i] = x_bin_list[i] // 2
                aux_list.append(1)   #per aggiungere 1 alla fine del while (è il resto)    
                aux_list.reverse()
                x_bin_list[i] = aux_list   
                
        print(x_bin_list)
    
        #qui converto in decimale
        
    
    else: 
        print("not a valid representation!")
    
    return 

conversion(x)

    
```

7\. Consider the integral of the semicircle of radius 1:
$$

$$

Alternatively we can use the Riemann definition of the integral:
$$
I=\lim_{N\to\infty} \sum_{k=1}^{N} h y_k 
$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.

(a) Write a programe to compute the integral with $N=100$. How does the result compares to the true value?
