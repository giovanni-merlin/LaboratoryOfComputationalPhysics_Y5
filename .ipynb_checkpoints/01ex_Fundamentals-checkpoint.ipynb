2\. Convert the following function into a pure function with no global variables or side effects

````python
x = 5
def f(alist):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)   #ok
print (alist) # alist has been changed! Not good!
````

````python
def f(alist, x):
    a = alist + [i for i in range(x)]
    return a

x = 5
alist = [1,2,3]
ans = f(alist,x)
print (ans)   #ok
print (alist) #alist has not been changed! Good!
````

````python
x = 5
def f(alist, x):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
a=alist[0:len(alist)]
ans = f(alist)
print (ans)   #ok
print (alist) # alist has not been changed! good!
````

````python
def f(alist, x):
    lista2 = [i for i in alist]
    for i in range(x):
         lista2.append(i)
    return lista2

alist = [1,2,3]
ans = f(alist,5)
print (ans)   #ok
print (alist) # alist has not been changed! good!
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
#n!=n*n-1!
#n=n-1*...*1

def fac1(n):
    """Factorial with recursion"""
    if n==0 or n==1:
        return 1 
    
    # recursion
    else:
        return n * fac1(n-1)
    
print([fac1(i) for i in range(10)])

def fac2(n):
    """Factorial without recursion"""
    
    a, b = 1, 1
    for i in range(1, n+1):
        print('a=', a, 'b=', b)
        a, b = a+1, a*b
        print('a=', a, 'b=', b, '\n' )
    return b    
    
print([fac2(i) for i in range(10)])

#GF
def fact(n):
    if(n==0):
        res = 1
    else :
        res = n
    for i in range(n-1,1,-1):
        res *= i
    return res
    
````

`I have 3 dogs. 
Tom is 6. 
Fletcher is 7. 
Larry is 9. 
And they're all mammals, of course. 
My dogs are not hungry.
`

```python
# Parent class
class Dog:

    # Class attribute
    species = 'mammal'

    # Initializer / Instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)

# Child class (inherits from Dog class)
class RussellTerrier(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)

# Child class (inherits from Dog class)
class Bulldog(Dog):
```
